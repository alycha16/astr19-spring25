```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 1, 100)

def exp_function(x):
    return np.exp(x)

y = exp_function(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x, y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.grid(True)

plt.savefig("awesome_plot.pdf")
plt.show()
```
