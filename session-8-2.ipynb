```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ar = np.linspace(0, 1, 100)

def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)

y_sin = sin_function(ar)
y_cos = cos_function(ar)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(ar, y_sin, color='lightgray')
ax1.set_title('sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.grid(True)

ax2.plot(ar, y_cos, color='black')
ax2.set_title('cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')
ax2.grid(True)

plt.tight_layout()
plt.savefig("sine_cosine_plot.pdf")
plt.show()
```
