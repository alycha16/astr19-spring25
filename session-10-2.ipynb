```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
original_array = np.linspace(0, 1, 1000)

np.random.shuffle(original_array)

random_numbers = original_array

plt.figure(figsize=(8,6))
plt.hist(random_numbers, bins=100, color='lightblue', edgecolor='pink')

plt.xlabel('Value')
plt.ylabel('Frequency')

plt.savefig('shuffle_histogram.pdf')
plt.show()
```
